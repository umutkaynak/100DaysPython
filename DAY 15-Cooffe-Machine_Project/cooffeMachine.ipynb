```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

tank = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}

jeton = {
    "Penny": 0.01,
    "Nickel": 0.05,
    "Dime": 0.10,
    "Quarter": 0.25
}
para = 0
bakiye = 0
devam = True


def report():
    print("su: ", tank["water"], "mg", "-", "süt: ", tank["milk"], "mg", "-", "kahve: ", tank["coffee"], "mg", "-",
          "para: ", bakiye, "cent")


def harcanan():
    if siparis == "latte":
        tank["coffee"] -= 24
        tank["milk"] -= 100
        tank["water"] -= 200
    elif siparis == "cappuccino":
        tank["coffee"] -= 24
        tank["milk"] -= 100
        tank["water"] -= 250
    elif siparis == "espresso":
        tank["coffee"] -= 18
        tank["water"] -= 50


while devam == True:
    siparis = input("ne arzu edersiniz (latte/cappucinno/espresso)").lower()
    if siparis == "of":
        break
    if siparis == "report":
        report()
        break

    if siparis == "latte" or siparis == "cappucinno":
        if MENU[siparis]["ingredients"]["water"] > tank["water"]:
            print("tankta yeterli miktar su yok.")
            break
        if MENU[siparis]["ingredients"]["coffee"] > tank["coffee"]:
            print("tankta yeterli miktarda kahve yok")
            break
        if MENU[siparis]["ingredients"]["milk"] > tank["milk"]:
            print("tankta yeterli miktarda süt yok")
            break
    else:
        if MENU[siparis]["ingredients"]["water"] > tank["water"]:
            print("tankta yeterli miktar su yok.")
            break
        if MENU[siparis]["ingredients"]["coffee"] > tank["coffee"]:
            print("tankta yeterli miktarda kahve yok")
            break

    para = 0
    Penny = int(input("kaç Penny var:"))
    Nickel = int(input("kaç Nickel var:"))
    Dime = int(input("kaç Dime var:"))
    Quarter = int(input("kaç Quarter var:"))

    para += (Penny * 0.01) + (Nickel * 0.05) + (Dime * 0.10) + (Quarter * 0.25)

    siparisFiyat = MENU[siparis]["cost"]

    if siparisFiyat > para:
        print(f"yeterli paranız yok, paranız:{para},sipariş fiyatı : {siparisFiyat} cent")
    elif siparisFiyat <= para:
        print(f"{siparis} hazır. afiyet olsun ☕")
        harcanan()
        if siparisFiyat < para:
            print((para - siparisFiyat), "cent para üstünüz")
            bakiye += siparisFiyat
```

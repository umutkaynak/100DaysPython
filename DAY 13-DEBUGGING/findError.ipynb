```python
number = int(input("Which number do you want to check?"))

if number % 2 == 0:
  print("This is an even number.")
else:
  print("This is an odd number.")
  
```

```python
year = input("Which year do you want to check?")

if year % 4 == 0:
  if year % 100 == 0:
    if year % 400 == 0:
      print("LEAP YEAR")
    else:
      print("NOT LEAP YEAR")
  else:
    print("LEAP YEAR")
else:
  print("NOT LEAP YEAR")
  

#GİRİLEN DEĞER STR OLUP İNT ARAMA YAPILMAKTA
```

```python
year = int(input("Which year do you want to check?"))

if year % 4 == 0:
  if year % 100 == 0:
    if year % 400 == 0:
      print("artık yıl")
    else:
      print("artık yıl değil.")
  else:
    print("artık yıl")
else:
  print("artık yıl değil")
  
```

```python
#FİZBUZZ 
#BİRDEN FAZLA HATA BLUNMAKTADIR.
for number in range(1, 101):
  if number % 3 == 0 or number % 5 == 0:
    print("FizzBuzz")
  if number % 3 == 0:
    print("Fizz")
  if number % 5 == 0:
    print("Buzz")
  else:
    print([number])
```

```python
for number in range(1, 101):
  if number % 3 == 0 and number % 5 == 0:
    print("FizzBuzz",[number])
    print("-")
  elif number % 3 == 0:
    print("Fizz",[number])
    print("*")
  elif number % 5 == 0:
    print("Buzz",[number])
    print("**")
  else:
    print(number)

#İLK OLARAK OR DEĞİL AND OLMALI YOKSA HE İKSİNDEDN BİRİ GERÇEKLEŞSEDE FİZBUZ YAZACAK
# if bloklarını değiştirdik çünkü sonraki else bir üstteki if bloğuna göre çalışıyordu.
```

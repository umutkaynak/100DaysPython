```python
row1 = ["⬜️","⬜️","⬜️"]
row2 = ["⬜️","⬜️","⬜️"]
row3 = ["⬜️","⬜️","⬜️"]
map = [row1, row2, row3]
print(f"{row1}\n{row2}\n{row3}")

position = input("hazineyi nereye koymak istersin ?")
column = int(position[0])
row = int(position[-1])
c1 = column-1
r1 = row-1
map[r1][c1] = "X"
print(f"{row1}\n{row2}\n{row3}")
```
